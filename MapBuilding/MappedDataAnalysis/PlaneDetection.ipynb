```python
import numpy as np
import json
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import axes3d
import math

%matplotlib inline
```

```python
data_arr = json.load(open('data.json'))
# data_arr.append(data_arr[0])
data = np.array(data_arr)
```

```python
data
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(data[:,0], data[:,1], data[:,2])
plt.show()
```

```python
pca = PCA(n_components=2)
collapsed = pca.fit_transform(data)

plt.plot(collapsed[:3,0], collapsed[:3,1])
plt.show()
```

```python
def distance(p1, p2):
    return (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2

def turn(p1, p2, p3):
    arg1 = math.atan((p2[1]-p1[1])/(p2[0]-p1[0]))
    arg2 = math.atan((p3[1]-p2[1])/(p3[0]-p2[0]))
    
    return arg2 - arg1
```
